# Analyse sentimental

```python
# Use a pipeline as a high-level helper
from transformers import pipeline
pipe = pipeline("text-classification", model="Peed911/french_sentiment_analysis")
```

```python

def transcribe_audio(audio):
  """
  Cette fonction permet de faire la transcription d'un fichier audio en français

  Args : Audio files

  return : texte
  """
  segments, info = model.transcribe(audio)
  for segment in segments:
    text = segment.text
  return text

def sentiment_analyse_text(text):
  """
  Cette fonction prend du texte et l'analyse pour détecter si c'est du sentiment positif ou négatif

  Args : str (texte)

  return : texte (expliquant le sentiment renvoyé par le modèle et la probabilité)

  """
  sentiment = pipe(text)
  final_result = {"transcription " : text, "sentiment" : sentiment[0]["label"]}
  return final_result

def inference(audio):
  """
  Cette fonction fait la prediction complete d'un audio vers un sentiment
  """
  text = transcribe_audio(audio)
  result = sentiment_analyse_text(text)
  print(result)
```

```python
## Mettez ici le ou les  liens de l'audio à tester
audio_path = ["/content/drive/MyDrive/DIT_DL2/exam_dl2/common_voice_fr_17299384.wav"]
```

```python
%%time
for audios in audio_path :
  inference(audios)
```
